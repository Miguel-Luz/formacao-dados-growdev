2 - A prefeitura de uma cidade fez uma pesquisa entre seus habitantes, coletando dados sobre o salário e número de filhos. A prefeitura deseja saber:

a) média do salário da população;

b) média do número de filhos;

c) maior salário;

d) percentual de pessoas com salário até R$2000,00.

Escreva um programa que receba as informações necessárias de 5 pessoas conforme a descrição e responda às questões a, b, c e d anteriores.


```python
def get_value(message):
     while True:
        try:
           return float(input(message))
        except ValueError:
         print("Somente números são aceitos para este campo.")
         
salarios = []
filhos = []

for i in range(1,6):
  salrio = get_value(f"Entrevistado 00{i} Qual o seu salário ? ")
  salarios.append(salrio)
  filho = get_value(f"Entrevistado 00{i} você tem quantos filhos ? ")
  filhos.append(filho)
  
  media_salario = sum(salarios) / len(salarios)
  media_filho = sum(filhos) / len(filhos)
  maior_salario = max(salarios)
  
print(f"A média de filhos é: {media_filho}")
print(f"A média de salários é: {media_salario}")
print(f"O maior salário é: {maior_salario}")
  
  
  
  
  
```

3 - Chico tem 1,50 metro e cresce 2 centímetros por ano, enquanto Zé tem 1,10 metro e cresce 3 centímetros por ano. Construa um programa que calcule e imprima quantos anos serão necessários para que Zé seja maior que Chico.

```python
altura_chico = 1.50
altura_ze = 1.10
crescimento_chico = 0.02
crescimento_ze = 0.03
anos = 0

while altura_ze <= altura_chico:
    altura_chico += crescimento_chico
    altura_ze += crescimento_ze
    anos += 1

print("Serão necessários", anos, "anos para Zé ser maior que Chico.")
```

4 - Em uma eleição presidencial existem dois candidatos. Os votos são informados através de códigos. Os dados utilizados para a contagem dos votos obedecem à seguinte codificação:

3 = voto nulo

Elabore um programa que leia o código de votação de 5 eleitores. Calcule e escreva:
total de votos para cada candidato

```python
def get_voto(message):
    while True :
     voto  = input(message)
     if voto in ['1','2','3','4']:
      return voto
     else: 
       print("Opção inválida")    
       
votos= []
for i in range(1,6):
  votos.append(get_voto(f"Eleitor número {i} qual o seu voto? ")) 

print(f"{votos.count('1')} voto(s) para o candidato 1")
print(f"{votos.count('2')} voto(s) para o candidato 2")
print(f"{votos.count('3')} voto(s) nulos")
print(f"{votos.count('4')} voto(s) em branco")  
  
  
  
  
      
```

5 - Escreva um programa que receba o nome de 10 pessoas e para cada pessoa, o lugar para o qual ela já viajou. Sendo que existem 3 possibilidades:

Rio de Janeiro

Bahia

Minas Gerais

Após, informar qual foi o destino mais visitado e qual o menos visitado.


```python
from collections import Counter

def get_destino(message, destinos):
 while True:
  destino = input(message)
  if destino in destinos:
      return destinos[destino]
  else:
    print("Opção inválida")

  viagens = []
  destinos_possiveis = {'1': 'Rio de Janeiro', '2': 'Bahia', '3': 'Minas Gerais'}

  for i in range(10):
    nome = input("Qual o seu nome")
    viagens.append(get_destino(f"Qual foi o seu destino {nome} 1)Rio de Janeiro 2)Bhaia 3)Minas Geraias", destinos_possiveis))
    
  contador = Counter(viagens)
  
  mais_visitado = max(contador, key=contador.get)

  print(f"O lugar mais visitado foi: {mais_visitado} ({contador[mais_visitado]} votos)")
  
        
  
  
```

9 - Escreva um programa que leia a idade e salário de 10 pessoas. Informe em seguida:
Qual é a média de idade entre as pessoas?
Quantas pessoas há por faixa etária, considerando:
jovens < 18
18 <= adultos < 60
idosos >= 60
Em seguida, mostre qual é a faixa etária que acumula o maior salário.


```python
def get_idade(message):
     while True:
        try:
           return int(input(message))
        except ValueError:
         print("Somente números inteiros são aceitos para este campo.")

def get_salario(message):
     while True:
        try:
           return float(input(message))
        except ValueError:
         print("Somente números são aceitos para este campo.")
      
def get_media_idade(populacao):
     total_idades = sum(map(lambda pessoa: pessoa['idade'], populacao))
     return total_idades / len(populacao)
  
def get_faixa_idade_abastada(dicionario):
     return max(dicionario, key=lambda x: x['total'])

def soma_salarios_faixa(label, faixa):
     return {'faixa':label,  'total': sum(map(lambda x: x['salario'], faixa))}
     
    
jovens  = []
adultos = []
idosos  = []
soma_salarios = []

for i in range(1,11):
    idade  =  get_idade(f"Entrevistado {i} qual a sua idade? ")
    salario =  get_salario(f"Entrevistado {i} qual o seu salário? ")
    
    dados = {'idade': idade, 'salario': salario}
    
    if idade >= 60:
     idosos.append(dados)   
    elif idade >= 18:
     adultos.append(dados)  
    else:
     jovens.append(dados)
     

populacao = sum([jovens, adultos, idosos], [])

soma_salarios.append(soma_salarios_faixa('jovens', jovens))   
soma_salarios.append(soma_salarios_faixa('adultos', adultos))
soma_salarios.append(soma_salarios_faixa('idosos', idosos))
maior_salario = get_faixa_idade_abastada(soma_salarios)

print(f"A média de idade da populção é {get_media_idade(populacao)} anos")
print(f"{len(jovens)} são jovens")
print(f"{len(adultos)} são adultos")
print(f"{len(idosos)} são idosos")
print(f"Os {maior_salario['faixa']} tem o maior salário totalizando: R${maior_salario['total']}" )    
```

11 - Crie um jogo em que o computador escolha um número aleatório entre 1 e 100, e o jogador tenta adivinhá-lo. O programa deve dar dicas se o número é maior ou menor do que o número digitado pelo jogador. O jogo continua até que o jogador acerte o número.

Exemplo de saída:

Bem-vindo ao jogo de adivinhação!

Tente adivinhar o número entre 1 e 100.

Digite sua tentativa: 50

O número é menor.

Digite sua tentativa: 75

O número é maior.

Digite sua tentativa: 63

O número é menor.

Digite sua tentativa: 68

Parabéns! Você acertou o número 68 em 4 tentativas!

```python
import random

def get_tentativa(message):
     while True:
        try:
           return int(input(message))
        except ValueError:
         print("Somente números são aceitos para este campo.")


numero = random.randint(1, 100)
    
while True:
    tentativa = get_tentativa("Tente adivinhar o número (entre 1 e 100): ")
        
    if tentativa == numero:
            print("Você acertou o número secreto:", tentativa)
            break
    elif tentativa < numero:
            print("O número é maior do que", tentativa)
    else:
            print("O número é menor do que", tentativa)
```

A fórmula para calcular o valor de um investimento após n anos com juros compostos é:

A=P×(1+r)^n


Onde:


A é o valor do investimento após n anos,

P é o principal (valor inicial do investimento),

r é a taxa de juros anual (em decimal), e

n é o número de anos.
