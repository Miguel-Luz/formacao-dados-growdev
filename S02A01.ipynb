1 - Conversão de graus Celsius para Fahrenheit – Crie um programa que converta graus
Celsius em Fahrenheit.

```python
# Versão de celsius_to_fahrenheit 

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

try:
    celsius = float(input("Digite a temperatura em graus Celsius: "))
    fahrenheit = celsius_to_fahrenheit(celsius)
    print(f"{celsius} graus Celsius é equivalente a {fahrenheit} graus Fahrenheit.")
except ValueError:
    print("Por favor, insira um valor numérico para a temperatura em graus Celsius.")
```

```python
# Fahrenheit para Celsius
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

try:
    fahrenheit = float(input("Digite a temperatura em graus Fahrenheit: "))
    celsius = fahrenheit_to_celsius(fahrenheit)
    print(f"{fahrenheit} graus Fahrenheit é equivalente a {celsius} graus Celsius.")
except ValueError:
    print("Por favor, insira um valor numérico para a temperatura em graus Fahrenheit.")
```

2- Escreva um programa que receba um número e escreva “Par” caso esse número seja par
e escreva “Impar” caso esse número seja impar.

```python
def verificar_par_ou_impar(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Ímpar"
    
try:
    numero = int(input("Digite um número: "))
    resultado = verificar_par_ou_impar(numero)
    print(resultado)
except ValueError:
    print("Por favor, insira um número inteiro.")
```

3 - Escreva um programa que receba dois números, exiba as opções:
1 - adição
2 - subtração
Então peça ao usuário para escolher uma das opções. Caso escolha a opção 1 o programa
deve realizar a soma dos dois números lidos e exibir. Caso escolha a opção 2 o programa
deve realizar a subtração dos dois números lidos e exibir.

```python
def adicao(num1, num2):
    print(f'{num1} + {num2} = {num1 + num2}')
    

def subtracao(num1, num2):
    print(f'{num1} - {num2} = {num1 - num2}')
     

try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))
        
        print("Escolha a operação:")
        print("1 - Adição")
        print("2 - Subtração")
        
        while True:
         opcao = int(input("Digite o número da operação desejada: "))
        
         if opcao == 1:
            adicao(num1, num2)
            break
         elif opcao == 2:
            subtracao(num1, num2)
            break
         else:
            print("Opção inválida. Por favor, escolha 1 para adição ou 2 para subtração.")
except ValueError:
    print("Por favor, insira números válidos.")
```

4 - Numa determinada escola, os critérios de aprovação são os seguintes:
- O aluno deve ter, no máximo, 25% de faltas;
- A nota final deve ser igual ou superior a 7,00.
Construa um programa para ler as notas que um aluno tirou nos 4 bimestres, o número total
de aulas e o número de faltas, mostrando ao final a situação do aluno como sendo
“Aprovado”, “Reprovado por Faltas” e “Reprovado por média”, considerando que a
reprovação por faltas se sobrepõe a reprovação por nota.

```python
def resolver_situacao(media, aulas, faltas):
    limite_faltas = 0.25 * aulas 
    if faltas > limite_faltas:
        return "Reprovado por Faltas"
    elif media < 7.0:
        return "Reprovado por Média"
    else:
        return "Aprovado"

def capturar_media():
    try:    
        notas = []
        for i in range(1, 5):
            while True:
             nota = float(input(f"Informe a nota do {i}º bimestre: "))
             if not 0 <= nota <= 10:
                print('Somente valores entre 0 e 10 são válidos')
             else:   
              notas.append(nota)
              break
        return sum(notas) / len(notas)
    except ValueError:
        print("Por favor, insira números válidos.")

media = capturar_media()
aulas_totais = int(input("Informe o número total de aulas: "))
faltas_totais = int(input("Informe o número de faltas: "))

situacao = resolver_situacao(media, aulas_totais, faltas_totais)
print(situacao) 
```

que um ano é bissexto quando for divisível por 4.

8 - Faça um programa que leia as duas notas parciais obtidas por um aluno numa
disciplina ao longo de um semestre, e calcule a sua média. A atribuição de conceitos
obedece à tabela abaixo:
Média de Aproveitamento
- Entre 9.0 e 10.0 Conceito - A
- Entre 7.5 e 8.9 - B
- Entre 6.0 e 7.4 - C
- Entre 4.0 e 5.9 - D
- Entre 0 e 3.9 - E

O programa deve mostrar na tela as notas, a média, o conceito correspondente e a
mensagem:
APROVADO se o conceito for A, B ou C.
REPROVADO se o conceito for D ou E.

```python
def get_float(message):
     while True:
        try:
           return float(input(message))
        except ValueError:
         print("Somente números são aceitos para este campo.")

notas = []         

for i in range(1, 3):
   while True:
      nota = get_float(f"Informe a {i}º nota do semestre: ")
      print(nota)
      if not 0 <= nota <= 10:
         print('Somente valores entre 0 e 10 são válidos')
      else:   
         notas.append(nota)
         break

media = sum(notas) / len(notas)         

if media >= 9:
   conceito = "A"
elif media >= 7.5:
   conceito = "B"
elif media >= 6:
   conceito = "C"
elif media >= 4:
   conceito = "D"
else:
   conceito = "E"

print(f"Nota 1: {notas[0]} Nota 2: {notas[1]} Media: {media}")   
print(f"Conceito: {conceito}")

if conceito in ['A', 'B', 'C']:
   print('APROVADO')
else:   
   print('REPROVADO')    

```

- salários entre R$ 280,00 e R$ 700,00: aumento de 15%
